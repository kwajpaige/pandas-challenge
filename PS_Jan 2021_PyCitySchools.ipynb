```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "PyCityschools/Resources/schools_complete.csv"
student_data_to_load = "PyCitySchools/Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Perform calculations
tot_num_schools = len(school_data_complete['School ID'])
tot_num_students = school_data_complete['student_name'].count()
tot_budget = sum(school_data_complete['budget'])
avg_math_score = school_data_complete['math_score'].mean()
avg_reading_score = school_data_complete['reading_score'].mean()

#passing math score/reading score = (count of students whose math score is greater than or equal to 70)/(total number of students) 
passmath_score_df=school_data_complete[school_data_complete['math_score']>=70]['student_name']
passmath_score=len(passmath_score_df)/tot_num_students*100

passreading_score_df = school_data_complete[school_data_complete['reading_score']>=70]['student_name']
passreading_score=len(passreading_score_df)/tot_num_students*100

# passing math and reading = (pass_math_score and pass_reading_score)/(total_num_students)
mar_df=school_data_complete[(school_data_complete['math_score']>=70) & (school_data_complete['reading_score']>=70)]['student_name']
mar=len(mar_df)/tot_num_students*100

#create dataframe
summary = {'Total Schools':[tot_num_schools],
            'Total Students': [tot_num_students], 
            'Total Budget': [tot_budget],
            'AVG Math Score': [avg_math_score], 
            'AVG Reading Score': [avg_reading_score],
            'Passing Math (>=70%)': [passmath_score], 
            'Passing Reading (>=70%)': [passreading_score],
            'Passing Reading and Math': [mar]}
district_summary = pd.DataFrame(summary)
district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#Groupby School Name
schools = school_data_complete.groupby(['school_name'])



#Use 'schools': determine values for other fields in table

school_type = schools['type'].first()
tot_students_per_school = schools['size'].first()
tot_budget_per_school = schools['budget'].first()
budget_per_student = tot_budget_per_school/tot_students_per_school
#AMS: Average Math Score, ARS: Average reding score, PMP: Passing Math Percentage, PRP: Passing reading Percentage
AMS = schools['math_score'].mean()
ARS = schools['reading_score'].mean()

PMP_df = school_data_complete[school_data_complete['math_score']>=70].groupby(['school_name']).size()
PMP = PMP_df/tot_students_per_school*100

PRP_df = school_data_complete[school_data_complete['reading_score']>=70].groupby(['school_name']).size()
PRP = PRP_df/tot_students_per_school*100

#PMRP: Passing Math and Reading Percentage
PMRP_df=school_data_complete[(school_data_complete['math_score']>=70) & (school_data_complete['reading_score']>=70)].groupby(['school_name']).size()
PMRP = PMRP_df/tot_students_per_school*100

#Create data frame
schools_summary = {'School Type': school_type,
                  'Total Students': tot_students_per_school,
                  'Total School Budget': tot_budget_per_school,
                  'Per Student Budget':budget_per_student, 
                  'Average Math Score':AMS,
                  'Average Reading Score':ARS,
                  '% Passing Math': PMP,
                  '% Passing Reading': PRP,
                  '% Overall Passing':PMRP}

schools_summary_df = pd.DataFrame(schools_summary)

schools_summary_df

```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
school_spending = schools_summary_df.copy()

#Create bins
bins = [0,585,630,645,680]
#Create label_groups
Label_groups = ['<$585', '$585 - $630', '$630 - $645', '$645 - $680']
school_spending["Per Student Budget"] = pd.to_numeric(school_spending["Per Student Budget"])
school_spending["Average Math Score"] = pd.to_numeric(school_spending["Average Math Score"])
school_spending["Average Reading Score"] = pd.to_numeric(school_spending["Average Reading Score"])
school_spending["% Passing Math"] = pd.to_numeric(school_spending["% Passing Math"])
school_spending["% Passing Reading"] = pd.to_numeric(school_spending["% Passing Reading"])
school_spending["% Overall Passing"] = pd.to_numeric(school_spending["% Overall Passing"])



#New Column for Spend ranges 


school_spending['Spending Range per Student'] = pd.cut(school_spending['Per Student Budget'], bins = bins, labels = Label_groups)


#Dataset for Scores by Spend ranges
school_spending = school_spending[['Average Math Score', 
                    'Average Reading Score', 
                    '% Passing Math', 
                    '% Passing Reading',
                    '% Overall Passing',
                  'Spending Range per Student']]

#Group by spend ranges and display dataset

Average_scores_by_spending = school_spending.groupby('Spending Range per Student').mean()

Average_scores_by_spending



```
